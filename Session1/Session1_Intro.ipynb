## ATLAS and the Large Hadron Collider

Beams of protons are accelerated around the Large Hadron Collider at speeds close to the speed of light, 
and brought to collision at 4 "interaction points" on the ring. One of the interaction points is at the 
center of the ATLAS detector. The collisions produce debris in the form of new particles which fly out at all angles.



Protons are collided in bunches - each bunch containing 100 billion protons (!) - at a rate of 30 million times a second. A so-called "bunch crossing" produces several $\mathcal{O}(10s)$ of proton-proton interactions (*see [here](https://www.lhc-closer.es/taking_a_closer_look_at_lhc/0.lhc_p_collisions) for more information on LHC proton bunches*). Occassionally, these interactions are spectacular head-on collisions and the full collision energy is used to produce a new system of massive particles that then decay to energetic lighter stable particles. These energetic stable particles are detected by the ATLAS detector: Their paths are tracked in the (inner) tracking system (if they have charge), and their energies measured when they are captured in the (outer) calorimeter system. Muons are not easily stopped and travel even further out to the (very outer) muon system, as is seen in this event display that captured a Z boson decaying to $\mu^+ + \mu^-$. Neutrinos are not tracked at all and will pass through all layers undetected. They can only be inferred by an "energy imbalance" measured in the event.

The [ATLAS Glossary](https://opendata.atlas.cern/docs/documentation/appendix/glossary) is a useful reference for further terms and concepts related to the ATLAS detector and events.

## Particle measurements in the ATLAS coordinate system

We "reconstruct" particles by combining measurements from different parts of the detector. The resulting particle will have an associated energy and direction and identified as a certain type, for example as a muon, electron, or hadronic shower initiated by a quark or gluon (reconstructed as a "jet").

The coordinate system of ATLAS is a right-handed coordinate system with the x-axis pointing towards the centre of the LHC tunnel, and the z-axis along the tunnel. The y-axis is slightly tilted with respect to vertical due to the general tilt of the tunnel. The azimuthal angle œï is measured from the x-axis, around the beam, and the polar angle $\theta$ is measured from the positive beam axis.
A particle will have the following properties: Energy E, $p_x$, $p_y$, $p_z$ (3-momentum vector). However, a particle will more commonly be described by its *tranvserse* momentum, $p_T$, $\phi$ and an angle related to $\theta$, $\eta$ (eta), known as *pseudorapidty*.

*Why specifically these coordinates*?

The **transerse momentum**, $p_T$, is the momentum in the transverse (x-y) plane, $p_T = p_y sin(\phi)$. Most proton-proton interactions are inelastic interactions in which by-products are of low energy and fly out in the same general direction as the initial protons, so close to the beam axis. An interaction with high transverse energy indicates a head-on collision and maximal energy transfer to the creation of a new system that decays isotropically, and is thus of more interest to us.

The **pseudorapidity** is defined as $\eta = ln(tan(\frac{\theta)}{2}))$. There are two reasons we prefer $\eta$ over $\theta$: (1) The $\eta$-distance between two particles is Lorentz invariant under boosts in the longitudinal direction, and (2) particle production is constant as a function of pseudorapidity. The former is an important feature, as the longitudinal boost in the lab frame is always different depending on the longitudinal momentum of the interacting quarks in the protons.
